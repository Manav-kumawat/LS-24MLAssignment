```python
#feature mapping
def map_feature(X1,X2):
  X1=np.atleast_1d(X1)
  X2=np.atleast_1d(X2)
  degree=6
  out=[]
  for i in range(1,degree+1):
    for j in range(i+1):
      out.append((X1**(i-j) * (X2**j)))
  return np.stack(out, axis=1)
```

```python

def compute_cost(X, y, w, b, lambda_=1):
    m = X.shape[0]
    z = np.dot(X, w) + b
    predictions = sigmoid(z)
    
    # Compute the cost
    cost = -(1 / m) * (np.dot(y, np.log(predictions)) + np.dot(1 - y, np.log(1 - predictions)))
    
    # Compute the regularization term
    reg_cost = (lambda_ / (2 * m)) * np.sum(w ** 2)
    
    # Total cost
    total_cost = cost + reg_cost
    
    return total_cost
```

```python
#compute gradient
def compute_gradient(X, y, w, b, lambda_=None):
#write your code

    m = len(y)
    
    # Compute predictions
    z = np.dot(X, w) + b
    h = sigmoid(z)
    
    # Compute gradients
    dj_dw = 1/m * np.dot(X.T, (h - y))
    dj_db = 1/m * np.sum(h - y)
    
    if lambda_ is not None:
        # Regularization for gradients if lambda_ is provided
        dj_dw += (lambda_ / m) * w

    return dj_db,dj_dw
```

```python
def plot_decision_boundary(w, b, X, y):
    # Plot data points
    plt.scatter(X[y == 1][:, 1], X[y == 1][:, 2], color='black', marker='+', label='Accepted')
    plt.scatter(X[y == 0][:, 1], X[y == 0][:, 2], color='yellow', marker='o', label='Rejected')

    # Define the decision boundary (x2 = - (w1*x1 + b) / w2)
    x1_min, x1_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    x2_min, x2_max = X[:, 2].min() - 1, X[:, 2].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.1),
                           np.arange(x2_min, x2_max, 0.1))
    Z = sigmoid(np.dot(map_feature(xx1.ravel(), xx2.ravel()), w) + b)
    Z = Z.reshape(xx1.shape)
    
    # Plot decision boundary
    plt.contour(xx1, xx2, Z, levels=[0.5], colors='blue')
    
    # Labels and legend
    plt.xlabel('Test 1')
    plt.ylabel('Test 2')
    plt.legend()
    plt.title('Decision Boundary')

    plt.show()

```
